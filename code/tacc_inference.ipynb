# Testing of TACC Models

```python
import json
from openai import OpenAI
import os

BASE_URL = "https://ai.tejas.tacc.utexas.edu/v1"   # endpoint
API_KEY = os.getenv("TACC_API_KEY")

MODELS = [
    # "Llama-4-Maverick-17B-128E-Instruct",
    # "Meta-Llama-3.3-70B-Instruct",
    "E5-Mistral-7B-Instruct"
]

client = OpenAI(api_key=API_KEY, base_url=BASE_URL)

results = {}

print("🚀 Testing model availability...\n")
for model in MODELS:
    print(f"Testing {model} ...", end=" ")
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": "Hello!"}],

        )
        msg = response.choices[0].message.content.strip()
        results[model] = {"status": "OK", "reply": msg}
        print("OK")
    except Exception as e:
        err = str(e)
        results[model] = {"status": "ERROR", "error_message": err}
        print("ERROR")

# ======= output =======
print("\nSummary:")
print(json.dumps(results, indent=2, ensure_ascii=False))
```
